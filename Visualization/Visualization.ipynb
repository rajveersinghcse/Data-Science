### Seaborn

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Univariate -> "One Variable" data visualization

```python
#strip plot
sns.stripplot(x='sex',y='tip',hue='day',data=tips, jitter=True)
plt.title('Strip Plot')
plt.ylabel('tip ($)')
plt.show
```

```python
#grouping with stripplot()
sns.stripplot(x='day',y='tip',hue='smoker',data=tips)
plt.ylabel('tips($)')
plt.show()
```

```python
sns.stripplot(x='sex',y='tip',hue='day',data=tips, jitter=False)
plt.title('Strip Plot')
plt.ylabel('tip ($)')
plt.show
```

```python
#swarm plot
plt.figure(figsize=(10,5))
sns.swarmplot(x='day',y='tip',data=tips, facecolor='black')
plt.ylabel('tips ($)')
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',data=tips, hue='sex')
plt.ylabel('tips ($)')
plt.show()
```

```python
#changing the orientation
plt.figure(figsize=(10,5))
sns.swarmplot(x='tip',y='day',data=tips, hue='sex')
plt.ylabel('tips ($)')
plt.show()
```

### Box and Violin plot

```python
plt.figure(figsize=(10,5))
sns.violinplot(x='day',y='tip',data=tips,color='lightgrey')
plt.ylabel('tips ($)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tips ($)')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tips ($)')

plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tips ($)')
plt.show()
```

```python
# Combine Plots

sns.violinplot(x='day',y='tip',data=tips, inner=None, color='lightgray')
sns.stripplot(x='day',y='tip',data=tips, size=4, jitter=True)
plt.ylabel('tips ($)')
plt.show()
```

### Bivariate -> "two variables" data visualization

```python
# joint plot
sns.jointplot(x='total_bill', y='tip',data=tips)
plt.show()
```

```python
# Density Plot

sns.jointplot(x='total_bill',y='tip', data=tips, kind='kde')
plt.show()
#contours are created representing various level of density of the data
#kernel density estimator - drawing normal distribution
```

```python
# pair plot
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

## Titanic Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
%matplotlib inline
```

### Loding dataset

```python
df = pd.read_csv('data/titanic_dataset.csv')
df.head()
```

```python
df.describe().T
```

```python
df.info()
```

```python
# Proportion of target (survived)
df.groupby('Survived')['PassengerId'].count()
```

```python
#Visual Exploration
f ,ax = plt.subplots(figsize=(7,5))
sns.boxplot(x='Survived',y='Age', data=df)
```

```python
f, ax = plt.subplots(figsize=(5,4))
sns.barplot(x='Sex',y='Survived',data=df)
```

```python
f, ax = plt.subplots(figsize=(10,5))
sns.boxplot(x='Sex',y='Age',hue='Survived', data=df)
```

```python
sns.barplot(x='Pclass',y='Survived',data=df)
```

```python
sns.barplot(x='Pclass',y='Survived',hue='Sex', data=df)
```

```python
sns.barplot(x='SibSp', y='Survived', data=df)
```

```python
sns.barplot(x='Parch', y='Survived', data=df)
```

```python
survived = df.loc[df['Survived']==1, "Age"].dropna()
sns.distplot(survived)
plt.title('Survived')
```

```python
not_survived = df.loc[df['Survived']==0, "Age"].dropna()
sns.distplot(not_survived)
plt.title('Not Survived')
```

```python
sns.pairplot(df.dropna());
```

```python
grid = sns.FacetGrid(df, col='Survived', row='Pclass', aspect=1.5)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend()
```

```python
plt.figure(figsize=(5,3))
sns.distplot(a=df[df['Embarked']=='C']['Survived'], bins=3, kde=False)
plt.title('Cherbourg')
plt.xticks([0,1])
plt.show()

plt.figure(figsize=(5,3))
sns.distplot(a=df[df['Embarked']=='Q']['Survived'], bins=3, kde=False)
plt.title('QueensTown')
plt.xticks([0,1])
plt.show()

plt.figure(figsize=(5,3))
sns.distplot(a=df[df['Embarked']=='S']['Survived'], bins=3, kde=False)
plt.title('Southampton')
plt.xticks([0,1])
plt.show()

```

```python
figbi, axesbi = plt.subplots(2,4, figsize=(18,10))
df.groupby('Pclass')['Survived'].mean().plot(kind='barh', ax=axesbi[0,0],xlim=[0,1])
df.groupby('SibSp')['Survived'].mean().plot(kind='barh', ax=axesbi[0,1],xlim=[0,1])
df.groupby('Parch')['Survived'].mean().plot(kind='barh', ax=axesbi[0,2],xlim=[0,1])
df.groupby('Sex')['Survived'].mean().plot(kind='barh', ax=axesbi[0,3],xlim=[0,1])
df.groupby('Embarked')['Survived'].mean().plot(kind='barh', ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x='Survived',y='Age',data=df, ax=axesbi[1,1])
sns.boxplot(x='Survived',y='Fare',data=df, ax=axesbi[1,2])
plt.show()
```
